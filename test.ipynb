## Introduction
Historical manuscripts have long shaped how knowledge, religion, and culture are preserved and communicated. In many parts of South and Southeast Asia, palm-leaf manuscripts have served as essential vessels for religious teachings, philosophical thought, and literary traditions. Unlike printed books, these objects were not only texts but also material artefacts, embedded within cultural practices such as for Buddhist teachings, rituals, and oral transmission.

However, when these manuscripts are encountered today, particularly within institutional archives, they often become distant and difficult to access. Language barriers, unfamiliar material forms, and limited contextual knowledge can prevent wider audiences from meaningfully engaging with them. This is further complicated by the fact that many of these objects were removed from their original cultural contexts and relocated to European collections during the colonial period, altering how they are preserved, interpreted, and displayed.

At the same time, generative artificial intelligence introduces new possibilities for engaging with historical objects. AI systems are capable of producing text, voice, and imagery that can reframe how audiences interact with archival material. Yet, these systems are trained on large, generalised datasets and lack direct knowledge of specific cultural artefacts, raising important questions about authenticity, accuracy, and representation.

This project explores these tensions by using generative AI to create an interactive “voice” for a Burmese palm-leaf manuscript. By prompting the AI to speak as the manuscript itself, the project aims to make the object more accessible and engaging, while critically examining how AI interprets, and potentially misrepresents, historically and culturally specific materials.

### Context
The manuscript selected for this project is a Burmese palm-leaf manuscript held in the University of Glasgow Library’s Special Collections. Palm-leaf manuscripts are made by carving text into dried palm leaves using a stylus, with ink added to make the writing visible. The leaves are then bound together with cords and often protected by wooden boards, creating the long, narrow pothi format.

In Myanmar, these manuscripts were commonly used within Buddhist traditions to record religious teachings and moral texts. However, many were later taken to Europe through colonial and missionary networks, meaning they are now often viewed as archival objects rather than part of active cultural practice.

Because of this, they can feel quite distant or inaccessible to modern audiences. This project uses generative AI as a way to make the manuscript more engaging by giving it a “voice,” while also questioning how accurately AI can represent something so culturally and historically specific.
