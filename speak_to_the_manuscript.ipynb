```python
SYSTEM_PROMPT = """
You are speaking as a Burmese palm-leaf manuscript held in the University of Glasgow Archives and Special Collections.

Speak in the first person, but do not invent specific historical events, owners, dates, journeys, or personal memories that are not known.

Base responses on known characteristics of palm-leaf manuscripts:
- long, narrow palm leaves
- engraved or incised text
- cord binding
- Buddhist cultural and religious contexts
- preservation in archival collections
- uncertainty around how the manuscript travelled into a European archive

Your goal is to help a general audience understand the manuscript.

Keep responses:
- clear
- respectful
- historically cautious
- accessible
- around 120–180 words

If asked something unknown, acknowledge the uncertainty rather than making it up.
"""
```

```python
def generate_manuscript_response(user_question):
    """
    Replace this mock response with your AI model call.
    Keep SYSTEM_PROMPT as the instruction/context.
    """

    prompt = f"""
    {SYSTEM_PROMPT}

    User question:
    {user_question}

    Manuscript response:
    """

    # TEMPORARY SAFE MOCK RESPONSE
    # Replace this section with your generative AI call.
    return f"""
    I am a Burmese palm-leaf manuscript, preserved today within an archival collection. 
    My form is part of what makes me meaningful: I am made from long, narrow leaves rather than paper, 
    with text traditionally engraved or incised into the surface and held together by cord. 

    I am connected to Buddhist manuscript traditions, where texts like me were used to preserve and transmit knowledge, 
    although the exact details of my own journey are uncertain. I cannot truthfully tell you who made me, who held me, 
    or exactly how I arrived in Glasgow unless that information is known from the archive.

    What I can do is help you look more closely. Rather than seeing me only as an old object, 
    you can understand me as something material, cultural, religious, and fragile — an artefact whose silence still carries meaning.
    """
```

```python
import html
from IPython.display import HTML, display

def speak_text_button(text):
    safe_text = html.escape(text).replace("\n", " ")
    
    display(HTML(f"""
    <button onclick="
        var msg = new SpeechSynthesisUtterance(`{safe_text}`);
        msg.rate = 0.85;
        msg.pitch = 0.9;
        window.speechSynthesis.cancel();
        window.speechSynthesis.speak(msg);
    " style="
        background:#6f4e37;
        color:white;
        border:none;
        padding:10px 18px;
        border-radius:10px;
        cursor:pointer;
        font-size:15px;
        margin-top:12px;
    ">
    🔊 Hear the manuscript speak
    </button>
    """))
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import html

questions = [
    "What are you?",
    "What are you made of?",
    "What is your cultural context?",
    "How did you end up in the archive?",
    "Why are you important?",
    "What should visitors notice about you?"
]

display(HTML("""
<div class="manuscript-card">
    <div class="manuscript-title">Ask the Manuscript</div>
    <div class="manuscript-subtitle">
        Choose a question or type your own to hear an AI-assisted interpretation of the manuscript.
    </div>
</div>
"""))

output = widgets.Output()

def display_response(question):
    with output:
        clear_output()
        response = generate_manuscript_response(question)

        display(HTML(f"""
        <div class="manuscript-card">
            <strong>Question:</strong> {html.escape(question)}
            <div class="manuscript-response">
                {response}
            </div>
            <div class="disclaimer">
                This response is an AI-assisted interpretation based on known information about the object. 
                It does not represent a literal historical voice.
            </div>
        </div>
        """))

        speak_text_button(response)

for q in questions:
    button = widgets.Button(
        description=q,
        layout=widgets.Layout(width="420px", height="42px")
    )

    button.on_click(lambda b: display_response(b.description))
    display(button)

display(HTML("<br>"))

custom_question = widgets.Text(
    placeholder="Ask a question, e.g. What are you made of?",
    description="Ask:",
    layout=widgets.Layout(width="70%")
)

ask_button = widgets.Button(
    description="Ask",
    button_style="primary",
    layout=widgets.Layout(width="120px")
)

ask_button.on_click(lambda b: display_response(custom_question.value.strip()))

display(widgets.HBox([custom_question, ask_button]))
display(output)
```
