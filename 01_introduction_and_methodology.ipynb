## Giving a voice to a silent object
Archival objects often sit quietly behind glass or within catalogues, preserved but rarely experienced. While they hold cultural, historical, and material significance, engaging with them can feel distant, especially when their language, form, and context are unfamiliar.

This project explores how AI can be used to reinterpret and communicate archival objects for a general audience. It focuses on the relationship between human input and AI-generated outputs, particularly how meaning is shaped through design decisions rather than generated automatically.

The project centres on a historical manuscript and experiments with different ways of presenting it through AI-generated text and audio. I approach AI as a collaborative tool, where interpretation is guided, edited, and refined through human input.

Throughout the project, I worked closely with my peer support group to test ideas, gather feedback, and refine outputs. These discussions played an important role in shaping both the direction of the project and the final outcomes.

## Encountering the Object
The object selected for this project is a [Burmese palm-leaf manuscript](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf) held in the University of Glasgow Library’s Special Collections. [Palm-leaf manuscripts](https://www.gla.ac.uk/myglasgow/archivespecialcollections/discover/specialcollectionsa-z/palm-leafandpothi/) are a traditional form of book used across South and Southeast Asia. They are made by carving text into dried palm leaves using a stylus, before applying ink to make the writing visible. The leaves are then bound together with cords and protected by wooden boards, forming the long narrow pothi format.

One thing I noticed when looking closely at the manuscript, was how rounded the script appeared. After doing some research, I found that this is closely linked to the material itself. Palm leaves are relatively fragile and can crack under pressure, so scripts developed [rounded letterforms](https://www.brandeis.edu/library/archives/essays/special-collections/burmese-texts.html) that are easier to inscribe without damaging the surface. This means the visual style of the text is directly shaped by the material it is written on, linking form, technique, and language together.

What makes this object particularly interesting is the contrast between its original context and its current setting within a Western archive through [colonial and missionary networks](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf).. While the manuscript carries cultural and spiritual signifance, its journey into the collection is not fully documented, introducing an element of uncertainty around its history

This project does not attempt to reconstruct a fictional narrative of that journey. Instead, it focuses on how the manuscript can be interpreted and presented today, using AI as a tool to explore different perspectives while acknowledging the limits of what is known.

## Design Approach
The design approach for this project is experimental and iterative. Rather than producing a single fixed outcome from the beginning, I developed a series of AI-generated outputs and refined them over time.

The project explores different formats, including first-person narratives and audio narration, to understand how the manuscript can be experienced in different ways. These formats were chosen to make the object more accessible and engaging, while also allowing space to question how interpretation is constructed.

As discussed with my peer support group, using a combination of text and audio helped create a more immersive experience, but also raised questions about how authoritative the outputs might appear.

## Methodology
This project uses a prompt-based approach to interact with a generative AI language model. By structuring prompts in different ways, the model produces varied interpretations of the manuscript, allowing the exploration of how meaning is shaped through Human-AI collaboration.

Workflow:

1. Design a prompt to guide the AI (e.g focusing on material form, purpose, or context)
2. Generate a response using the AI model
3. Analyse the output in relation to tone, accuracy, and interpretation
4. Refine the prompt based on feedback and evaluation
5. Repeat the process to compare and improve results

This iterative workflow is central to the project. As discussed with my peer support group, small changes in prompt design can significantly affect how the manuscript is represented, particularly in terms of tone and perceived authority.

## Human - AI Collaboration
This project is built around a collaborative process between human input and generative AI output. A large language model (LLM) is used to generate responses based on prompts that ask it to interpret the manuscript.

While the AI produces the text, I shape the outcome through decisions around prompting, framing, and selection. As discussed with my peer support group, even small changes in wording can significantly affect tone and meaning.

From a machine learning perspective, the model is trained on large-scale text datasets and generates responses by predicting likely word sequences. It does not have direct knowledge of the manuscript or access to archival records. Instead, it relies on generalised patterns from its training data.

This introduces limitations. The outputs may:

- Lack cultural or historical specificity
- Include inaccuracies or assumptions
- Present speculative interpretations as factual

For this reason, the AI is treated as an interpretive tool rather than an authoritative source. The value of this process lies not only in what the AI produces, but in analysing how those outputs are shaped.

## Access and Ethics
The manuscript is held in the University of Glasgow Library’s Archives and Special Collections. While it is publicly accessible, it requires booking and careful handling to ensure preservation.
This project explores how AI can extend access by presenting the manuscript through digital formats such as text and audio. However, this also raises ethical considerations.
 
### Copyright and Data Use
All materials used must comply with copyright and licensing regulations. Archival content may have restrictions, and permission is required where necessary. AI systems are trained on large datasets that are not fully transparent, raising questions about data ownership and consent.
 
### Cultural Representation
The manuscript originates from a Buddhist and Southeast Asian context. AI does not have direct cultural understanding, and its outputs are shaped by generalised data. As discussed with my peer support group, this creates risks of oversimplification or misrepresentation.
To address this, the project avoids unverifiable claims and acknowledges uncertainty.
 
### Authority and Authenticity
Presenting the manuscript in a first-person voice may suggest authenticity. However, this voice is constructed through AI and prompting. The outputs are therefore treated as interpretive rather than factual, and transparency is maintained throughout.
