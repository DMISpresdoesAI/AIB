# Notebook 3: Reflection and Evaluation
This notebook critically reflects on the outcomes of the project, evaluating the effectiveness of the methodology, the role of human–AI collaboration, and the ethical considerations involved.

## Evaluation of Project Objectives
This project set out to explore how AI can be used to interpret archival objects for a general audience, while also examining how much human input actually shapes AI-generated content.

Overall, I think the project was successful in showing that AI can make archival material more engaging, especially through first-person narration and audio. 
However, the experiments also showed a clear trade-off between engagement and accuracy. In earlier outputs, particularly in Experiment 2, the AI produced a more narrative-driven response that felt convincing but wasn't grounded in the specific manuscript. This reflects a known issue with languge models, where they can generate fluent but unreliable information (Bender et al., 2021)

Through this iteration, I refined the prompts to produce outputs that remained engaging while avoiding fictionalisation and acknowledging uncertainty. This directly addressed the project objectives, particularly in showing how design choices influence meaning and audience perception.

## Human-AI Collaboration
A key insight from this project is that AI outputs are heavily dependent on the human input, whether that the simple prompt, or the information that it is trained on. Small changes in prompt design resulted in significantly different interpretations, highlighting the importance of direction and refinement. 

As discussed with my peer support group, the most effective outputs were those where the AI was given clear constraints. This supports the idea of human-AI collaboration as a co-creative process, where meaning is shaped through interaction rather than generated independently (Boden, 2016).

## Limitations and Ethics
This project also highlighted several limitations. The language model does not have access to the specific manuscript (other than the images uploaded for refinement), meaning that the outputs were based on generalised knowledge rather than object-specific data. This can lead to inaccuracies and a false sense of authority.

The use of a first-person narration and audio further amplifies this, making interpretations feel more authentic than they are. This raises ethical concerns around representation, particularly as the manuscript originates from a specific cultural context that the AI does not seem to fully understand. 

While I addressed this by refining the prompts and acknowledging uncertainty, these limitations still remain an important consideration when using AI in cultural heritage contexts.

## The Key Takeaway
The most important takeaway from this project is that that the effectiveness of AI depends most on how it is used rather than on the tool itself.

Prompt design, iteration, and critical evaluation were central to producing meaningful outputs. At the same time, multimodal elements such as audio increased engagement while also increasing the risk of perceived authority.

## Final Reflection
Overall, this projects demonstrates both the potential and limitations of using AI to engage audiences with cultural heritage.

AI can mae archival material more accessible and engaging, however it requires careful direction to avoid misrepresentation or misinterpretation. The process of working with AI was not just about generating content, but about shaping and questioning it, reinforcing the importance of human involvement in AI-driven interpretation.
