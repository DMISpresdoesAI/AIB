# Giving Voice to a Burmese Palm-Leaf Manuscript  
### A Human–AI Collaborative Approach to the Arts and Humanities

## Introduction

This project explores how Human–AI collaboration can be used to reinterpret and communicate cultural heritage objects to a general audience. It focuses on a Burmese palm-leaf manuscript held in the [University of Glasgow Archives and Special Collections](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf).

[Palm-leaf manuscripts](https://www.gla.ac.uk/myglasgow/archivespecialcollections/discover/specialcollectionsa-z/palm-leafandpothi/) are traditionally made from long, narrow leaves with engraved text, often bound together with cords. They are commonly associated with Buddhist textual traditions and were used to preserve religious and cultural knowledge. While the manuscript holds significant cultural value, it can be difficult for a general audience to engage with due to language barriers, unfamiliar material form, and limited contextual information.

This project addresses this challenge by using AI to present the manuscript in a more accessible and engaging way, while maintaining respect for its material form, cultural context, and historical uncertainty.

## The Challenge

Archival objects are often preserved for research and conservation, but not always designed for public engagement. In the case of this manuscript, several factors contribute to this difficulty:

- The text is not easily readable to those unfamiliar with the language  
- The physical format is unfamiliar to many audiences  
- The physical object is very fragile and requires care. As a result, booking an appointment with the University archives is required to view the object.
- Limited contextual information makes interpretation challenging  

As a result, the manuscript can feel distant and difficult to connect with, despite its cultural and historical significance.

## Project Objective

The aim of this project is to make the manuscript more accessible by exploring how AI can support interpretive engagement.

This is achieved by:
- Using AI to generate descriptive, first-person responses  
- Presenting the manuscript as if it can "speak" about itself  
- Creating an interactive system where users can engage through guided questions  

Rather than generating fictional narratives, the project focuses on reframing existing knowledge in a way that is more approachable for non-expert audiences.

## Methodology

This project adopts a Human–AI collaborative approach.

The human role involves:
- Designing prompts  
- Defining constraints to prevent inaccurate or fictional outputs  
- Selecting and refining responses  

The AI is used to:
- Generate responses based on these prompts  
- Explore different ways of expressing information  

To maintain accuracy and integrity, all responses are constrained to:
- Focus on the manuscript’s material form  
- Reflect its cultural and religious context  
- Acknowledge uncertainty where information is unknown  
- Avoid inventing specific historical events or narratives  

This ensures that the output remains grounded while still being engaging.

```python
def generate_manuscript_response(user_question):

    responses = {
        "What are you?": """I am a Burmese palm-leaf manuscript, now held in the University of Glasgow Archives and Special Collections. I do not look like what you might expect from a book. I am not bound with a spine or made of paper. Instead, I exist as a stack of long, narrow leaves, each one carefully cut and shaped, then threaded together through small holes and held between wooden boards. My form follows a tradition known as the pothi format, used across South and Southeast Asia, where text is arranged in horizontal lines across each strip.""",

        "What are you made of?": """I am made from palm leaves that were once part of a living tree. They were processed—cut, dried, and smoothed—so they could carry text. But my words are not written in ink in the usual sense. They are carved into my surface. If you look closely, you can see the fine incisions that form each character. After carving, a dark substance was applied so the letters would stand out. Over time, my edges have darkened, and my surface shows signs of handling. I am both resilient and delicate, shaped by use and by age.""",

        "What is your cultural context?": """I come from a cultural context closely connected to Buddhist knowledge and practice in Myanmar. Manuscripts like me were used to record teachings, commentaries, and other important texts. I was not designed to be read quickly or silently. Instead, I would have been handled with care, each leaf turned in sequence, the text followed slowly, often spoken aloud. My physical form shapes the way I am read—it encourages attention, rhythm, and patience.""",

        "How did you end up in the archive?": """I am now held within the University of Glasgow Archives and Special Collections, but the details of my journey here are not fully known. Like many archival objects, I have been separated from my original context, and the circumstances of my movement are uncertain. What can be said is that I am now preserved as part of a collection, where I am studied and cared for. This uncertainty is part of my story, reflecting broader histories of collection and preservation.""",

        "Why are you important?": """I am important not only for the text I carry but also for my material form. As a palm-leaf manuscript, I represent a method of recording and preserving knowledge that differs from more familiar book formats. I am connected to Buddhist traditions and reflect cultural practices of inscription and preservation. Even without full knowledge of my specific history, I offer insight into broader systems of knowledge, material culture, and archival preservation.""",

        "What should visitors notice about you?": """Visitors should notice my physical form: the long, narrow leaves, the engraved text, and the way I am bound together. These features are not decorative but functional, reflecting the methods used to create and preserve manuscripts like me. You might also consider the marks of wear and the texture of the material, which speak to my age and use. While I may appear unfamiliar, these details reveal the cultural and material context from which I originate."""
    }

    return responses.get(user_question, "I am a manuscript with a complex and partially unknown history.")
```

```python
import html
from IPython.display import HTML, display

def speak_text_button(text):
    safe_text = html.escape(text).replace("\n", " ")
    
    display(HTML(f"""
    <button onclick="
        var msg = new SpeechSynthesisUtterance(`{safe_text}`);
        msg.rate = 0.85;
        msg.pitch = 0.9;
        window.speechSynthesis.cancel();
        window.speechSynthesis.speak(msg);
    " style="
        background:#6f4e37;
        color:white;
        border:none;
        padding:10px 18px;
        border-radius:10px;
        cursor:pointer;
        font-size:15px;
        margin-top:12px;
    ">
    🔊 Hear the manuscript speak
    </button>
    """))
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import html

questions = [
    "What are you?",
    "What are you made of?",
    "What is your cultural context?",
    "How did you end up in the archive?",
    "Why are you important?",
    "What should visitors notice about you?"
]

display(HTML("""
<div class="manuscript-card">
    <div class="manuscript-title">Ask the Manuscript</div>
    <div class="manuscript-subtitle">
        Choose a question or type your own to hear an AI-assisted interpretation of the manuscript.
    </div>
</div>
"""))

output = widgets.Output()

def display_response(question):
    with output:
        clear_output()
        response = generate_manuscript_response(question)

        display(HTML(f"""
        <div class="manuscript-card">
            <strong>Question:</strong> {html.escape(question)}
            <div class="manuscript-response">
                {response}
            </div>
            <div class="disclaimer">
                This response is an AI-assisted interpretation based on known information about the object. 
                It does not represent a literal historical voice.
            </div>
        </div>
        """))

        speak_text_button(response)

for q in questions:
    button = widgets.Button(
        description=q,
        layout=widgets.Layout(width="420px", height="42px")
    )

    button.on_click(lambda b: display_response(b.description))
    display(button)

display(HTML("<br>"))
display(output)
```

This highlights the role of Human–AI collaboration as a creative process, where meaning is shaped through design decisions rather than generated automatically.

## Ethical Considerations

This project recognises the importance of representing cultural heritage objects responsibly.

The manuscript is not treated as a fictional narrator, and care is taken to avoid inventing historical details or personal narratives. Instead, responses are grounded in known characteristics and openly acknowledge uncertainty where information is unavailable.

AI is used as a tool to reinterpret and communicate existing knowledge, rather than to create new or speculative histories. This ensures that the object’s cultural and historical integrity is maintained.
