## Methodology Overview

This project uses a prompt-based approach to interact with a generative AI language model. The model generates text in response to use inputs, allowing different interpretations of the manuscript to emerge depending on how prompts are structured.

## Tools and AI
The primary tool used in this project is ChatGPT which is a large language model (LLM). LLMs are trained on large-scake text datasets, including books, websites, and other written sources. Therefore, instead of retrieving factual information, the model will generate responses based on patterns and probabilities learned during training.

Through discussion with my peer support group, ChatGPT was identified as the most appropriate tool for this project. This is becuase text-based generation allows for a clearer exploration of how AI constructs meaning, making it easier to analyse the relationship between input (prompt) and output (response). Compared to image or audio generation tools, text outputs provide more accessible opportunities for critical reflection, particularly when evaluating tone, accuracy, and cultural representation.

While additional tools such as AI voice generators were considered to enhance accessibility, the project primarily focuses on text generation in order to maintain clarity in methodology and analysis. This decision reflects an intentional balance between creative experimentation and the ability to critically engage with how generative AI systems operate.

It is also important to recognise the limitations of the tool. ChatGPT is not specifically trained on Burmese palm-leaf manuscripts, meaning its outputs are shaped by generalised knowledge rather than domain-specific expertise. As a result, the responses generated in this project are treated as interpretive and speculative, instead of authoritative.

Rather than pages, these manuscripts consist of long, narrow strips of dried palm leaves, bound together between wooden covers. Text is not written with ink in the conventional sense, but carefully incised into the surface using a stylus, before soot or charcoal is applied to make the lettering visible.

These manuscripts were commonly used to document religious texts—particularly Buddhist teachings—as well as medicine, law, and literature. Often produced and preserved in monasteries, they reflect both the intellectual traditions and craftsmanship of Burmese culture. Today, palm-leaf manuscripts are valued as important historical artefacts, offering insight into pre-modern knowledge systems and methods of record-keeping.

## EXPERIMENT 2: PERSONA BASED PROMPT

### PROMPT:
You are a Burmese palm-leaf manuscript. Describe your life, purpose, and journey.

### OUTPUT:
> *I was born from a tree.*
>
> Long before I carried words, I was a palm leaf, swaying in the humid air of Myanmar. I felt the rhythm of wind and rain, the slow passage of seasons. Then one day, I was cut, boiled, and dried—transformed from something living into something that could remember.
>
> A careful hand shaped me. I was smoothed, trimmed into a long, narrow strip, and stacked with others like me. Then came the moment that gave me purpose: a scribe pressed a sharp stylus into my surface. Each stroke carved meaning into my body. I did not feel pain—only the quiet weight of responsibility. The words etched into me were not ordinary. They were teachings, stories, knowledge meant to endure.
>
> Dark soot was rubbed across my surface, settling into the grooves of each letter. Suddenly, I could speak.
>
> I lived most of my early life in a monastery. I was handled with care, read aloud in low, steady voices. I listened as monks traced the lines inscribed upon me, passing knowledge from one generation to the next. I was not meant to be owned—I was meant to be shared, preserved, and respected.
>
> Time moved differently for me. I did not age in years, but in touch. The more I was read, the more I fulfilled my purpose. Yet I was also fragile. Too much light, too much moisture, too many careless hands—and I could fade, crack, disappear.
>
> Eventually, my journey changed. I left the monastery and entered a new kind of space—an archive. Here, I am no longer read in the same way. I am protected, studied, sometimes digitised. Fewer hands touch me now, but more eyes may see me than ever before.
>
> I have become both object and evidence.
>
> I remember a time when knowledge was slow—when each word required intention, effort, and care. Now I exist in a world where words are generated instantly, endlessly. Still, I remain.
>
> I am not just what I say.
> I am how I was made.
> I am what has survived.

### ANALYSIS
The AI-generated response adopts a first-person narrative voice, presenting the manuscript as a speaking subject with a life history. This approach is highly effective in making the object more engaging and accessible, aligning with the project's aim of giving the manuscript a "voice". The use of sensory and temporal language, such as "I was born from a tree" and "I lived most of my early life in a monastery," creates a compelling narrative that encourages the reader to connect with the object on a more personal level.

The response also incorporates several accurate material and cultural detials, including the transformation of palm leaves into writing surfaces, the use of a stylus to inscribe text, and the role of moasteries in preserving and transmitting knowledge. This suggests that the model is able to draw on generalised knowledge about palm-leaf manuscripts and Buddhist practices.

However, despite its fluency and coherence, the response is not grounded in the specific manuscript being studied. The narrative introduces a continuous and singular life story, which simplifies what is likely a more complex and uncertain history. For example, the transition from monastery to archive is presented as a linear journey, whereas in reality, the movement of such objects often includes multiple stages, actors, and historical contexts, particuarly within colonial collecting practices.

Additionally, the emotional tone of the response demonstrates how the model prioritises narrative engagement over historical facts and accuracy. The manuscript is anthropomorphised, given throughts and experiences that enhance readability but do not reflect the nature of the object itself.

This highlights a key characteristic of generative AI: its ability to produce convincing and meaningful narratives without direct knowledge or understanding of the subject. The output is shaped by patterns in the training data, which creates an illusion of authenticity despite being onstructed by generalised and probablistic associations.

### Analysis
The redined prompt produces a noticeably different type of response compared to the eariler persona-based output. While the model still adopts a first-person voice, the tone is more restrained and descriptive, focusing on material characteristics and contextual information rather than a fully imagined narrative.

The response demonstrates a stronger alignment with the physical and cultural features of the manuscript. It accurately refers to key elements such as the incised text, cord binding, wooden covers, and the manuscript's association with Buddhist knowledge and practice. This suggests that providing structured guidance within the prompt helps the model generate outputs that are more grounded in generalised factual knowledge.

Additionally, the response does acknowledge uncertainty, particularly in relation to the manuscript's journey into a European archive. This reflects the constraints introduced in the prompt, showing that the model can be guided not only in what it describes, but in how it frames knowledge and limitation.

However, it is clear that the output still remaines constructed from learned patterns rather than direct knowledge. The description is still generalised and could apply to many similar objects. The model does not verify information or access archival records. 

This experiment demonstrates the impact of prompt design on AI behaviour. By introducing constraints, the output becomes more critically aware and less prone to speculative storytelling. This therefore highlights the role of the human not only in generating outputs, but in shaping the boundaries of what the AI claims to know.
