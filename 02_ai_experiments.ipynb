## Tools and AI
The primary tool used in this project is ChatGPT which is a large language model (LLM). LLMs are trained on large-scake text datasets, including books, websites, and other written sources. Therefore, instead of retrieving factual information, the model will generate responses based on patterns and probabilities learned during training.

Through discussion with my peer support group, ChatGPT was identified as the most appropriate tool for this project. This is becuase text-based generation allows for a clearer exploration of how AI constructs meaning, making it easier to analyse the relationship between input (prompt) and output (response). Compared to image or audio generation tools, text outputs provide more accessible opportunities for critical reflection, particularly when evaluating tone, accuracy, and cultural representation.

While additional tools such as AI voice generators were considered to enhance accessibility, the project primarily focuses on text generation in order to maintain clarity in methodology and analysis. This decision reflects an intentional balance between creative experimentation and the ability to critically engage with how generative AI systems operate.

It is also important to recognise the limitations of the tool. ChatGPT is not specifically trained on Burmese palm-leaf manuscripts, meaning its outputs are shaped by generalised knowledge rather than domain-specific expertise. As a result, the responses generated in this project are treated as interpretive and speculative, instead of authoritative.
