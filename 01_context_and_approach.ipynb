## Introduction: Giving a voice to a silent object
Archival objects often sit quietly behind glass or within catalogues, preserved but rarely experienced. While they hold rich cultural, historical, and material significance, engaging with them can feel distant, especially when their language, form, and context are unfamiliar to certain audiences.

This project explores how AI can be used to reinterpret and communicate archival objects for a general audience. It focuses on the relationship between human input and AI generated outputs, particularly how meaning is shaped through design decisions rather than generated automatically.

The project centres on a historical manuscript and experiments with different ways of presenting it through AI-generated text and audio. This work approaches AI as a collaborative tool, where interpretation is guided, edited, and refined through human input.

Throughout the project, I worked closely with my peer support group to test ideas, gather feedback, and refine outputs to create a good user experience overall. These discussions played an important role in shaping both the direction of the project, and how the final outcomes were presented.

## Encountering the Object
The object selected for this project is a [Burmese palm-leaf manuscript](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf) held in the University of Glasgow Library’s Special Collections. [Palm-leaf manuscripts](https://www.gla.ac.uk/myglasgow/archivespecialcollections/discover/specialcollectionsa-z/palm-leafandpothi/) are a traditional form of book used across South and Southeast Asia. They are made by carving text into dried palm leaves using a stylus, before applying ink to make the writing visible. The leaves are then bound together with cords and protected by wooden boards, forming the long narrow pothi format.

One thing I noticed when looking closely at the manuscript, was how rounded the script appeared. After doing some research, I found that this is closely linked to the material itself. Palm leaves are relatively fragile and can crack under pressure, so scripts developed [rounded letterforms](https://www.brandeis.edu/library/archives/essays/special-collections/burmese-texts.html) that are easier to inscribe without damaging the surface. This means the visual style of the text is directly shaped by the material it is written on, linking form, technique, and language together.

What makes this object particularly interesting is the contrast between its original context and its current setting within a Western archive through [colonial and missionary networks](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf).. While the manuscript carries cultural and spiritual signifance, its journey into the collection is not fully documented, introducing an element of uncertainty around its history

This project does not attempt to reconstruct a fictional narrative of that journey. Instead, it focuses on how the manuscript can be interpreted and presented today, using AI as a tool to explore different perspectives while acknowledging the limits of what is known.

The design approach for this project is experimental and iterative. Rather than producing a single fixed outcome from the beginning, I developed a series of AI-generated outputs and refined them over time.

The project explores different formats, including first-person narratives and audio narration, to understand how the manuscript can be experienced in different ways. These formats were chosen to make the object more accessible and engaging, while also allowing space to question how interpretation is constructed.

As discussed with my peer support group, using a combination of text and audio helped create a more immersive experience, but also raised questions about how authoritative the outputs might appear.

## Human - AI Collaboration

This project is built around a collaborative process between human input and generative AI output. A large language model (LLM) is used to generate responses based on prompts that ask it to take on the perspective of the manuscript.
 
The workflow follows an iterative process:
1.	A prompt is designed by the human (e.g. asking the manuscript to describe its material form or purpose)
2.	The AI generates a response based on patterns learned from its training data
3.	The output is analysed, evaluated, and compared against the project objectives
4.	The prompt is refined to adjust tone, perspective, or level of interpretation
 
While the AI produces the text, the human plays a central role in shaping the outcome. Human decisions around specific phrasing, framing, and selection directly influence how the manuscript is represented. As discussed with my peer support group, even small changes in prompt wording significantly affects the tone and perceived authority of the responses.
 
From a machine learning perspective, the LLM used in this project (ChatGPT) is trained on large-scale text datasets and operates by predicting likely word sequences based on input prompts. It does not have direct knowledge of the specific manuscript, nor access to archival records. Instead, it generates responses based on generalised linguistic and cultural patterns present in its training data.
 
This introduces important limitations. While the outputs may appear coherent and convincing, they may:
•	Lack cultural and historical specificity
•	Introduce inaccuracies or assumptions
•	Present speculative interpretations as factual
 
For this reason, the AI is not treated as an authoritative voice, but as a speculative and interpretive tool. The value of this collaboration lies not only in the generated outputs, but in critically examining how those outputs are formed and how human decisions shape them.

## Access and Ethics

The manuscript used in this project is held in the University of Glasgow Library’s Archives and Special Collections. While access to the physical object is available to the public, it requires booking an appointment and following handling guidelines to ensure its preservation.
 
This project explores ways of increasing access through digital and AI-mediated interpretation. By generating text and audio outputs, the manuscript can be experienced by a wider audience without requiring direct physical access.
However, the use of generative AI raises several ethical considerations.
 
### Copyright and Data Use
Any materials used in this project, including images of the manuscript, must comply with copyright and licensing regulations. Archival materials may have specific restrictions on reproduction and reuse, and permission must be obtained where required.
The AI tools used are trained on large datasets that are not fully transparent. As a result, it is not always possible to determine the exact sources of the training data. This raises broader questions about data ownership and consent in AI systems.
 
### Cultural Representation
The manuscript originates from a Buddhist and Southeast Asian context. While AI can generate interpretations of the object, it does not have direct cultural understanding. Its outputs are shaped by generalised training data rather than specific knowledge of Burmese traditions.
As discussed with my peer support group, this creates a risk of:
•	Oversimplification of cultural meaning
•	Misrepresentation of religious context
•	Imposing Western narrative structures onto non-Western objects
To address this, the project avoids making unverifiable claims and acknowledges uncertainty where appropriate.
 
### Authority and Authenticity
Presenting the manuscript in a first-person voice can create a sense of authenticity and authority. However, this voice is constructed through AI and human prompting rather than originating from the object itself.
This project therefore treats AI-generated outputs as interpretive rather than factual. Maintaining transparency about this process is essential to avoid misleading the audience.
