## Introduction: Giving a voice to a silent object
Archival objects often sit quietly behind glass or within catalogues, preserved but rarely experienced. While they hold rich cultural, historical, and material significance, engaging with them can feel distant, especially when their language, form, and context are unfamiliar to certain audiences.

This project explores a different way of encountering such an object. Instead of presenting it through description alone, it asks: what if the manuscript could speak?

Using generative AI, this project experiments with creating a voice for a Burmese palm-leaf manuscript, allowing it to narrate its own history, purpose, and journey. In doing so, the project aims to make the object more accessible and engaging, while also questioning how accurately AI can represent culturally specific and historically situated artefacts.

Rather than treating AI as a tool for producing definitive interpretations, this project positions it as a collaborator that generates opportunities. However, it also introduces limitations, biases, and gaps. The result is not just a new way of presenting the manuscript, but a critical exploration of how meaning is shaped through both human and AI interpretation.

## Encountering the Object
The object selected for this project is a [Burmese palm-leaf manuscript](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf) held in the University of Glasgow Library’s Special Collections. [Palm-leaf manuscripts](https://www.gla.ac.uk/myglasgow/archivespecialcollections/discover/specialcollectionsa-z/palm-leafandpothi/) are a traditional book form, created by carving text into dried palm leaves using a stylus and adding ink to make the writing visible. The leaves are then bound together with cords and often protected by wooden boards, forming the long, narrow pothi format.

The physical material of the manuscript also shaped how it was written. Because palm leaves are relatively fragile and can crack under pressure, scripts developed [rounded letterforms](https://www.brandeis.edu/library/archives/essays/special-collections/burmese-texts.html) that were easier to inscribe without damaging the surface. This means the visual style of the text is directly influenced by the material it is written on, linking form, technique, and language together.

In Myanmar, these manuscripts were commonly used within Buddhist traditions to record religious teachings and moral texts. They were not just texts, but objects embedded in cultural and spiritual practices — handled, read aloud, and used within specific contexts.

However, many of these manuscripts were later removed from their original environments and brought into European collections through [colonial and missionary networks](https://www.gla.ac.uk/media/Media_1144973_smxx.pdf). As a result, they are now often encountered as archival artefacts rather than living cultural objects.

This shift creates a gap between the manuscript and its audience. Without knowledge of the language, context, or material form, it can be difficult to engage with the object beyond its appearance. This project attempts to bridge that gap by reintroducing a sense of narrative and presence — while also recognising that any reconstruction is inevitably partial and mediated.

## Project Objectives
This project aims to explore how generative AI can be used to create new forms of engagement with archival material, while critically examining its limitations.

The key objectives are:

- To use generative AI to create an interactive "voice" for the manuscript, making it more accessible to a wider audience
- To investigate how AI-generated outputs shape the interpretation of historical and cultural objects
- To critically evaluate the accuracy, assumptions, and limitations of AI when dealing with culturally specific material
- To demonstrate the role of human interpretation in guiding, refining, and contextualising AI-generated content

By combining creative experimentation with critical reflection, the project introduces a new perspective on how archival objects can be experienced in a digital context.

## Human - AI Collaboration

This project is built around a collaborative process between human input and generative AI output. A large language model (LLM) is used to generate responses based on prompts that ask it to take on the perspective of the manuscript.

The workflow follows a simple but interative process:
1. A prompt is designed by the human (e.g. asking the manuscript to describe its history)
2. The AI generates a response based on patterns learned from its training data
3. The output is analysed, evalyated, and refined
4. The prompt is adjusted to explore different perspectives or improve relevance

While the AI produces the text, the human plays a central role in shaping the outcome by deciding how the manuscript is framed, what exact questions are asked, and how the responses are interpreted.

It is also important to recognise the limitations of this collaboration. The model used in this project is trained on large-scale, general text datasets rather than specialised collections of Burmese manuscripts. This means that while it can generate fluent and engaging responses, it may lack cultural specificity, introduce inaccuracies, or rely on generatlised assumptions.

As a result, the AI is not treated as an authoritive voice, but as a speculative one. The value of this process lies not only in what the AI produces, but in critically examining how those outputs are formed, where they fall short, and what this reveals about the use of AI in the arts and humanities.

## Access and Ethics

The manuscript used in this project is held in the University of Glasgow Library's Archives and Special Collections. Access to the physical object is available to the public, but requires booking an appointment in advance and following appropriate archival handling guidelines. This ensure the preservation of the manuscript while still allowing it to be studied and engaged with.

This project involves the use of generative AI, as such, it is important to consider the issues of copyright and data use. Any material used as input for AI tools must either be in the public domain ir used with appropriate permission from the copyright holder. This is particularly relevant when working with archival material, where ownership and reproduction rights may vary depnding on the object. 

The use of gnerative AI does not replace the need for ethical responsibility; instead, it require additional awareness of how data is sourced, processed, and reproduced.

More broadly, this project also raises questions about cultural representation. While AI can generate engaging interpretations, it does not have direct access to the cultural, linguistic, or historical context of the manuscript. As a result, the outputs must be treated critically, with the understanding that they are shaped by generalised training data rather than specific cultural knowledge.
